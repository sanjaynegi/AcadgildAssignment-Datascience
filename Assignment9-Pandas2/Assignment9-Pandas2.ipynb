```python
import pandas as pd

df_Data= pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')


print(df_Data.info())


df_Data.head(2)
```

```python
# 1. Delete unnamed columns

df_Data_DeletdCol= df_Data.drop(['Unnamed: 0'],axis=1)

df_Data_DeletdCol.head()
```

```python
# 2. Show the distribution of male and female

round(df_Data['Gender'].value_counts(normalize=True)*100,2)
```

```python
# 3. Show the top 5 most preferred names

df_Data[["Name","Count"]].groupby('Name').sum().sort_values("Count",ascending=0).head(5)
```

```python
# 4. What is the median name occurence in the dataset

df_Data_median_Name=df_Data[["Name","Count"]].groupby('Name').sum()

print("Median name occurence in the dataset \n")
df_Data_median_Name[df_Data_median_Name['Count'] == df_Data_median_Name['Count'].median()]
```

```python
# 5. Distribution of male and female born count by states

df_Data_Gender_Distribution= df_Data[['State','Gender','Count']].set_index(['State', 'Gender'])


df_Data_Gender_Distribution.groupby(['State','Gender']).sum()
```
